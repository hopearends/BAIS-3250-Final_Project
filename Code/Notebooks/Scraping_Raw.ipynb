# Scraping Data

#### NFL Team Attendance From 2010 - 2023. This data was scraped in conjunction with a key statistics csv file found on Kaggle to analyze how attendance trends have changed over time along with what ket statistics effect attendance numbers.#

#### The data ws scrapped from this url https://www.espn.com/nfl/attendance/_/year/20XX (with the XX depending on what year)

```python
# import required libraries

import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import random
```

```python
# Set up the Selenium WebDriver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

try:
    # Define the URL for NFL attendance
    url = 'https://www.espn.com/nfl/attendance/_/year/2023'
    driver.get(url)

    # Allow the page to load
    time.sleep(random.uniform(5, 10))

    # Define the column names as you provided
    column_names = [
        'rank', 'team', 'home_gms', 'home_total', 'home_avg', 
        'road_games', 'road_total', 'road_avg', 
        'overall_gms', 'overall_total', 'overall_avg'
    ]
    
    # Find rows with classes starting with 'evenrow team-' or 'oddrow team-'
    even_rows = driver.find_elements(By.CSS_SELECTOR, "[class^='evenrow team-']")
    odd_rows = driver.find_elements(By.CSS_SELECTOR, "[class^='oddrow team-']")

    # Combine both lists
    all_rows = even_rows + odd_rows

    # Extract text from each row and combine into a list
    data = []
    for row in all_rows:
        cells = row.find_elements(By.TAG_NAME, 'td')
        row_data = [cell.text.strip() for cell in cells]
        data.append(row_data)

        # Create a DataFrame from the combined data and add column names
        attendance_2023 = pd.DataFrame(data, columns=column_names)
    
finally:
    driver.quit()
```

```python
# Create a list of the attendance DataFrames

attendance_dfs = [attendance_2023, attendance_2022, attendance_2021, attendance_2019, attendance_2018, attendance_2017, attendance_2016, attendance_2015, attendance_2014, attendance_2013, attendance_2012, attendance_2011, attendance_2010]
```

```python
# change the columns names to be easier to read

columns_to_convert = [
    'home_gms', 'home_total', 'home_avg',
    'road_games', 'road_total', 'road_avg',
    'overall_gms', 'overall_total', 'overall_avg'
]

def convert_columns_to_numeric(df, columns):
    for col in columns:
        if col in df.columns:
            df[col] = df[col].str.replace(',', '', regex=False)
            # Convert the column to numeric
            df[col] = pd.to_numeric(df[col], errors='coerce')

# Convert columns for each DataFrame (2023, 2022, 2021)
for df in attendance_dfs:
    df['rank'] = pd.to_numeric(df['rank'], errors='coerce')

#check the first and the last dataframs dtype for verfication
attendance_2023.dtypes
attendance_2010.dtypes
```

```python
# Loop through each DataFrame and convert the 'rank' column to numeric

for df in attendance_dfs:
    df['rank'] = pd.to_numeric(df['rank'], errors='coerce')
```

```python
# Reindex that dataframes

for df in attendance_dfs:
    df.reset_index(drop=True, inplace=True)
```

```python
# display the first few values of the first the last dataframe to ensure verfication

print(f'2023 Attendance:')
display(attendance_2023.head(10))

print(f'\n\n2010 Attendance:')
display(attendance_2010.head(10))
```

```python
# List of years excluding 2020
years = [2023, 2022, 2021, 2019, 2018, 2017, 2016, 2015, 2014, 2013, 2012, 2011, 2010]

# Loop through each DataFrame and rename the columns, adding a year column
for df, year in zip(attendance_dfs, years):
    # Rename each column to append the year to the column name, except 'team'
    df.columns = [f'{col}_{year}' if col != 'team' else col for col in df.columns]

    # Add a new column 'year' that contains the year value from the column names
    df['year'] = year

# Verify the column names
print(attendance_dfs[1].columns)
print(attendance_dfs[1].head())
```

```python
# Initialize the merged_data with the first DataFrame
merged_data = attendance_dfs[0]

# Merge the rest of the DataFrames
for df in attendance_dfs[1:]:
    merged_data = pd.merge(merged_data, df, on='team', how='inner', suffixes=('', f'_{df["year"].iloc[0]}'))

# Display the merged DataFrame
print("Merged Data:")
display(merged_data.head())
```

```python
# Generate column names for dropping from 2023 to 2010 for both 'road' and 'home'
years = range(2023, 2009, -1)  # From 2023 to 2010
columns_to_drop = []

# Generate the column names for 'road' and 'home' for each year
for year in years:
    columns_to_drop.extend([f'road_games_{year}', f'road_total_{year}', f'road_avg_{year}',
                            f'home_gms_{year}', f'home_total_{year}', f'home_avg_{year}'])

# Drop the columns from the merged DataFrame
merged_data = merged_data.drop(columns=columns_to_drop, errors='ignore')

# Optionally, display the first few rows to check the result
print("Merged Data after dropping columns:")
display(merged_data.head())
```

```python
# Generate the list of 'rank' columns to drop for each year
columns_to_drop_rank = [f'rank_{year}' for year in years]

# Drop the 'rank' columns
merged_data = merged_data.drop(columns=columns_to_drop_rank, errors='ignore')

#display the new dataframe
display(merged_data.head())
```

```python
# put raw data into csv

merged_data.to_csv("raw_data.csv", index=False, encoding= 'utf-8')
```

```python
# Initialize the renaming dictionary
rename_dict = {}

# Loop through the years to create the column renaming mappings
for year in years:
    rename_dict[f'overall_total_{year}'] = f'fan_att_{year}'
    rename_dict[f'overall_avg_{year}'] = f'avg_fan_att_{year}'

# Apply the renaming
merged_data = merged_data.rename(columns=rename_dict)

# Verify the changes by displaying the updated columns
print("Updated columns after renaming:")
print(merged_data.columns)
```
